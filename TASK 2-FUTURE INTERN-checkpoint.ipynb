```python
import numpy as np #importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

#### merge of dataset , cleaning, removing outliers

```python
df1 = pd.read_csv(r'C:\Users\dell\Downloads\titanic\gender_submission.csv')
df2 = pd.read_csv(r'C:\Users\dell\Downloads\titanic\test.csv')
df3 = pd.read_csv(r'C:\Users\dell\Downloads\titanic\train.csv')
```

```python
merge_data = pd.merge(df1, df2, on= 'PassengerId', how= 'inner')
```

```python
all_df = pd.concat([df3, merge_data], ignore_index=True)
```

```python
pd.isnull(all_df).sum()
```

```python
all_df['Age'].fillna(all_df['Age'].median(), inplace=True)
```

```python
all_df['Fare'].fillna(all_df['Fare'].median(), inplace=True)
```

```python
all_df['Embarked'].fillna(all_df['Embarked'].mode()[0], inplace=True)
```

```python
all_df['Cabin'].fillna('Unknown', inplace=True)
```

```python
pd.isnull(all_df).sum()
```

```python
all_df.info()
```

```python
all_df.describe()
```

```python
sns.boxplot(all_df['Fare'])
```

```python
Q1 = all_df['Fare'].quantile(0.25)
Q3 = all_df['Fare'].quantile(0.75)
IQR = Q3-Q1

lower_bound = Q1 - 1.5 * IQR    
upper_bound = Q3 + 1.5 * IQR 

print("Q1:", Q1)
print("Q3:", Q3)
print("IQR:", IQR)
print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)


outliers = all_df[(all_df['Fare'] < lower_bound) | (all_df['Fare'] > upper_bound)]
print("Outliers:\n", outliers)
```

```python
new_df = all_df.copy() #capping the outliers
new_df.loc[(new_df['Fare'] > upper_bound), 'Fare'] = upper_bound
new_df.loc[(new_df['Fare'] < lower_bound), 'Fare'] = lower_bound


print(new_df['Fare'].min(), new_df['Fare'].max())
```

```python
sns.boxplot(new_df['Fare'])
```

```python
plt.figure(figsize=(10, 5)) #COMPARING OLD AND NEW
sns.histplot(all_df['Fare'], kde=True, color='grey', label='Old Data', bins=30)
sns.histplot(new_df['Fare'], kde=True, color='orange', label='New Data', bins=30)
plt.title('Comparison of Fare Distributions')
plt.legend()
plt.show()
```

# calculate summary statistics

```python
columns = new_df.select_dtypes(include=['number'])

summary_stats = columns.agg(['mean', 'median', 'std']).T #".T"transpose:use to read the data esily

summary_stats['mode'] = columns.mode().iloc[0] #".iloc[0]"for first mode of the list

print(summary_stats)
```

```python
summary_stats[['mean', 'median', 'std']].plot(kind='line', figsize=(10, 6), marker='o')
plt.title('Summary Statistics Line Chart')
plt.ylabel('Values')
plt.xlabel('Columns')
plt.grid()
plt.show()
```
