```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
%matplotlib inline
import seaborn as sns
import matplotlib as mpl
```

```python
df = pd.read_csv(r'C:\Users\dell\Downloads\archive (27)\Iris.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
numerical_columns = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']
for column in numerical_columns:
    plt.figure(figsize=(6, 4))
    sns.histplot(data=df, x=column, kde=True, bins=20, color='blue')
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```

## From this we can see that sepal length and width of flowers are kinnda wide and petal width and length are kind of small in size . 

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Species', palette='pastel')
plt.title('Count of Each Species')
plt.xlabel('Species')
plt.ylabel('Count')
plt.show()
```

## number of quantity of Iris species are same.

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(
    data=df, 
    x='PetalLengthCm', 
    y='PetalWidthCm', 
    hue='Species', 
    style='Species', 
    palette='Set2', 
    s=100
)
plt.title('Petal Length and Petal Width')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.legend(title='Species')
plt.show()

#BAR GRAPH FOR SPECIES   

species_avg = df.groupby('Species')[['PetalLengthCm', 'PetalWidthCm']].mean().reset_index()


species_avg.plot(
    x='Species', 
    kind='bar', 
    figsize=(8, 6), 
    color=['skyblue', 'orange'], 
    alpha=0.8
)


plt.title('Average Petal Length and Width by Species', fontsize=16)
plt.ylabel('Measurement (cm)', fontsize=12)
plt.xlabel('Species', fontsize=12)
plt.xticks(rotation=0) 
plt.legend(['Petal Length', 'Petal Width'], title='Attributes')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

## From the above graphs i found that in scatter plot it is clear in seen that Iris versicolor and Iris virginica have similar size where on the other side Iris setosa have difference in size in comparison to them even in bar graph it seen clearly that versicolon and virginica makes a huge difference from setosa. 

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(
    data=df, 
    x='SepalLengthCm', 
    y='SepalWidthCm', 
    hue='Species', 
    style='Species', 
    palette='Set1', 
    s=100
)
plt.title('Sepal Length vs. Sepal Width')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.legend(title='Species')
plt.show()

#BAR GRAPH OF THE ABOVE GRAPH FOR BETTER VISUALIZATION.

species_avg = df.groupby('Species')[['SepalLengthCm', 'SepalWidthCm']].mean().reset_index()


species_avg.plot(
    x='Species', 
    kind='bar', 
    figsize=(8, 6), 
    color=['lightpink', 'grey'], 
    alpha=0.8
)


plt.title('Average Petal Length and Width by Species', fontsize=16)
plt.ylabel('Measurement (cm)', fontsize=12)
plt.xlabel('Species', fontsize=12)
plt.xticks(rotation=0) 
plt.legend(['Sepal Length', 'Sepal Width'], title='Attributes')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

## From the above graphs i found that just like petal, sepal also found the same difference between virginica, versicolon and setosa but in bar graph we took as an average just like petal seen and an difference the did't make hug difference between them.
